### 5 number summary and box plot

max,

min,

median,

q1 25, 

q3 75 percentile,

IQR

```python
import numpy as np
```

```python
lst_marks=[54,87,43,99,86,21,67,43,87,90,65,78,97,43,56,76,75,82,53,4,88,1,65,55,67]
minimum,q1,median,q3,maximum=np.quantile(lst_marks,[0,0.25,0.5,0.75,1.0])
```

```python
print(len(lst_marks))
print(minimum,q1,median,q3,maximum)
```

```python
IQR=q3-q1
lower=q1-(1.5*IQR)
upper_fence=q3+(1.5*IQR)
```

```python
update_lst={x for x in lst_marks if x>lower and x<upper_fence}
```

```python
len(update_lst)
```

```python
lower
```

```python
update_lst
```

```python
import seaborn as sns
sns.boxenplot(lst_marks)
```

```python
sns.boxenplot(update_lst)
```
