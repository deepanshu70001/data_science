```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#creatoing a dataset
from sklearn.datasets import make_classification
```

```python
X,y=make_classification(n_samples=10000,n_features=10,n_classes=2,random_state=42)
```

```python
pd.DataFrame(X)
```

```python
y
```

```python
#train test split
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.30,random_state=42)
```

```python
#model training
from sklearn.linear_model import LogisticRegression
```

```python
logic=LogisticRegression()
```

```python
logic.fit(X_train,y_train)
```

```python
y_pred=logic.predict(X_test)
```

```python
y_pred
```

```python
#preformance matrics
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
```

```python
sc=accuracy_score(y_test,y_pred)
print(sc)
cm=confusion_matrix(y_test,y_pred)
print(cm)
```

```python
print(classification_report(y_test,y_pred))
```

```python
### hyper parameter tuning
```

```python
model=LogisticRegression()
```

```python
penalty=['l1','l2','elasticney']
c_value=[100,10,0.1,0.01,1]
solver=['newton-cg','saga','sag','liblinear','lbfgs']
```

```python
params=dict(penalty=penalty,C=c_value,solver=solver)
```

```python
##grid search cv
from sklearn.model_selection import GridSearchCV,StratifiedKFold
cv=StratifiedKFold()
grid=GridSearchCV(estimator=model,param_grid=params,scoring='accuracy',cv=cv)
```

```python
print(grid)
```

```python
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
###randomised search cv
from sklearn.model_selection import RandomizedSearchCV
```

```python
randoms=RandomizedSearchCV(estimator=model,param_distributions=params,scoring='accuracy',cv=cv)
```

```python
randoms.fit(X_train,y_train)
```

```python
randoms.best_score_
```
