```python
#importing lib
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#loading data set
from sklearn.datasets import load_iris
iris=load_iris()
```

```python
iris
```

```python
print(iris['DESCR'])
```

```python
print(iris['target'])
```

```python
#independent features
X=pd.DataFrame(iris['data'],columns=iris['feature_names'])
```

```python
X.head()
```

```python
y=iris['target']
```

```python
#train test split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=10,test_size=0.2)
```

```python
#no need of standardization in decision tree
#apply decision tree
from sklearn.tree import DecisionTreeClassifier
dec_tree=DecisionTreeClassifier(max_depth=2)
dec_tree.fit(X_train,y_train)
```

```python
y_pred=dec_tree.predict(X_test)
```

```python
#visualize tree
from sklearn import tree
plt.Figure(figsize=(16,9))
tree.plot_tree(dec_tree,filled=True)
```

```python
#preventing overfitting by post or pre pruming
#by using post pruming
 #in this we create a tree then we cut the botton tree for puming
#better for small tree
#by giving max depth to  dec_tree=DecisionTreeClassifier(max_depth=2)
```

```python
#preformance metrice
from sklearn.metrics import classification_report,confusion_matrix
print("confusion matrix\n",confusion_matrix(y_test,y_pred))
print("report",classification_report(y_test,y_pred))
```

```python
params={'criterion':['gini', 'entropy', 'log_loss'],
        'splitter':['best','random'],
        'max_depth':[1,2,3,4,5],
       'max_features':['sqrt','log2','auto'],    
}
```

```python
from sklearn.model_selection import GridSearchCV
tree_model=DecisionTreeClassifier()
grid=GridSearchCV(tree_model,param_grid=params,cv=5,scoring='accuracy')
grid.fit(X_train,y_train)
```
