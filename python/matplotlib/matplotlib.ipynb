```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5,6]
y=[1,4,9,16,25,36]
#a line plot
plt.plot(x,y)
```

```python
#customised line plot
plt.plot(x,y,color='red',linestyle='-.',marker='o')
plt.grid(True)
plt.title("plot 1")
```

```python
#multiple plot
x=[1,2,3,4,5,6]
y=[1,4,9,16,25,36]
y1=[2,4,6,8,10,12]
plt.figure(figsize=(12,6))

plt.subplot(3,3,1)
plt.plot(x,y,color='green')
plt.title("p1")

plt.subplot(3,3,2)
plt.plot(x,y1,color='red')
```

```python
#bar plot
cat=['a','b','c','d','e']
val=[3,5,6,7,8]
plt.bar(cat,val,color='red')
plt.title("bar")
```

```python
#histogream
data=[4,5,4,1,6,8,3,1,6,5,7,9,8,6,4,2,1,2,5,6,8,7,5,4,2,1,2,9,9]
plt.hist(data,bins=5,color='purple',edgecolor='black')
```

```python
#scatter plot
x=[1,2,3,4,5,6,7]
y=[3,4,5,6,7,8,9]

plt.scatter(x,y,color='red',marker='x')
```
