```python
import seaborn as sns
```

```python
#basic ploting
t=sns.load_dataset('tips')
```

```python
t
```

```python
#scatter plot
import matplotlib.pyplot as plt

sns.scatterplot(x='total_bill',y='tip',data=t)
```

```python
#line plot
sns.lineplot(x='size',y='total_bill',data=t)
```

```python
sns.barplot(x='day',y='total_bill',data=t)
```

```python
#box plot
sns.boxplot(x='day',y='total_bill',data=t)
```

```python
sns.histplot(t['total_bill'],bins=5,kde=False)
```

```python
#kde
sns.kdeplot(t['total_bill'])
```

```python
#pair plot
sns.pairplot(t)
```

```python
#heat map
cor=t[['total_bill','tip','size']].corr()
sns.heatmap(cor)
```
